```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
hw08 = pd.read_csv('hw8.csv', header = None)

X = hw08.iloc[:,0:500]
Y = hw08.iloc[:,500:1000]
theta = hw08.iloc[:,-1]
w = np.random.normal(0,1,(len(X),len(X)))
list1 = np.zeros(len(X))
```

```python
for i in range(len(X)):
    approx = 0
    for j in range(len(X)):
        wi = w[j,:]
        approx = approx + (max(0,np.dot(wi,X.values[i,:]))* max(0,np.dot(wi,Y.values[i,:])))
    list1[i] = (approx/len(X))

plt.plot(theta,list1,'.',label = 'kapprox')
plt.plot(theta,(np.sin(theta)+(np.pi-theta)*np.cos(theta))/(2*np.pi),'r',label = 'kac')
plt.title("The value of kapprox and kac")
```
